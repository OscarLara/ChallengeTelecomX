## Extracción
Empezamos por la extracción de datos directamente desde el enlace

```python
# Comenzamos importando todas las librerías necesarias.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Nos conectamos a la fuente de datos directa:
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"
datos = pd.read_json(url)
# Vemos las primeras filas del DataFrame.
datos.head()
```

Nos pudimmos conectar sin problema a los datos originales y cargar un DataFrame.<br>
### Conoce los datos

## Transformación
Ya que todas nuestras bases están cargadas, podemos empezar a transformar los datos según nos convenga, para ello, primero voy a concatenar la información.

```python
# De los datos originales, solo queremos conservar la columna "Churn"
# Queremos concatenar las tablas horizontalmente, éso es el eje 1.
df = pd.concat([datos["Churn"], customer, phone, internet, account], axis=1)
df.head()
```

```python
# Verificamos la información del nuevo DataFrame.
df.info()
```

Aquí corroboramos que ninguna columna tiene datos nulos. Ahora verficaremos los datos numéricos, si todos están enlistados como tal y si son del tipo correcto.<br>
* SeniorCitizen: int64, es correcto.
* tenure: int64, es correcto.
* Charges.Monthly: float64, es correcto.
* Charges.Total: object, es falso. Hay que cambiar el tipo de dato a float64, ya que es un valor numérico con decimales.

Toca verificar que los datos de cada columna son lo esperado.

```python
# Verificamos si en la columna "Churn", todos los valores son "Yes" o "No".
df["Churn"].unique()
```

Notamos que hay 3 tipos de datos, "No", "Yes" y  ''. Debemos sustituir '' por np.NaN y quitarlo de la base ya que no nos sirve pues no sabesmos si son cancelaciones o no.

```python
# Convertimos nuestros valores vacíos a NaN para poder trabajar con ellos.
df["Churn"] = df["Churn"].replace('', np.nan)
df["Churn"].unique()
```

```python
# Retiramos las filas con valores NaN en la columna "Churn".
df.dropna(subset=["Churn"], inplace=True)
df["Churn"].unique()
```

```python
# Verificamos cuántas entradas nos quedan.
df.info()
```

Nos quedamos con 7043 entradas que sabemos si cancelaron o no su servicio

```python
# Intentamos transformar el dato a float64.
df["Charges.Total"].astype("float64")
```

El error nos indica que hay entradas con valores que no se pueden convertir en numero, en este caso fue un espacio vacío.

```python
df["Charges.Total"].replace(' ', np.nan).astype("float64")
```

```python
# La celda pasada funcionó, por lo que podemos guardar la variable y volvemos a verificar el DataFrame.
df["Charges.Total"] = df["Charges.Total"].replace(' ', np.nan).astype("float64")
df.info()
```

```python
df.head()
```

Nuestro DataFrame está listo para empezar el análisis.

```python
df.describe()
```

A partir de este punto, podemos manipular los datos para conseguir la información que queremos.<br>
Como lo que nos interesa es saber por qué la gente está cancelando, crearemos un DataFrame con solo los datos de las personas que han cancelado.

```python
# Filtramos los valores de las personas que sí cancelaron el servicio.
churn = df[df["Churn"]=="Yes"]
churn.info()
```

Observamos que tenemos 1869 personas que han cancelado de los 7043 datos totales. Ésto representa un $26.53\%$ de la base.<br>
Como ya filtramos las personas que sí cancelaron, podemos eliminar la columna de "Churn".


```python
churn =churn.drop(columns=["Churn"])
churn.head()
```

Notamos que el índice mantiene la numeración previa, como éste es un nuevo DF, le pondremos sus propios índices.

```python
churn = churn.reset_index()
churn = churn.drop(columns=["index"])
churn.head()
```

## Carga y Análisis
Podemos hacer un análisis por cada columna, sin embargo ya que la empresa da servicios, nos enfocaremos en los servicios ofrecidos.

¿Cuántos clientes se están yendo?

```python
plt.figure(figsize=(6, 4))
ax = sns.countplot(data=df, x="Churn")
ax.bar_label(ax.containers[0], fontsize=8, fontweight="bold")
plt.title("Distribución de Evasión (Churn)")
plt.show()

print("Porcentaje de Cancelación:")
print((df["Churn"].value_counts(normalize=True)*100).map('{:.2f}%'.format))
```

Tenemos un porcentaje de cancelación del $26.54\%$

¿Afecta el tipo de contrato?

```python
plt.figure(figsize=(10,6))
ax = sns.countplot(data=df, x="Contract", hue="Churn")
ax.bar_label(ax.containers[1], fontsize=8, fontweight="bold")
plt.title("Distribución de Evasión por Tipo de Contrato")
plt.show()
```

Podemos notar que donde más cancelaciones se presentan es en los contratos mes a mes.<br>
También inferimos que es donde más contratos hay. sin embargo, la proporcion de cancelaciones sigue siendo mucho mayor en cancelaciones mes a mes.

¿Afecta el tipo de contrato a la fuga?


```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="Churn", y="tenure")
plt.title("Antigüedad del Cliente por Evasión")
plt.show()
```

Podemos observar que entre mayor antigüedad se tenga, menos cancelaciones hay.

```python
cols_cat = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents', 
    'PhoneService', 'MultipleLines', 'InternetService', 
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
    'TechSupport', 'StreamingTV', 'StreamingMovies', 
    'PaperlessBilling', 'PaymentMethod'
]

# Configuramos el tamaño de la figura (varias filas y columnas)
plt.figure(figsize=(20, 30))

for i, col in enumerate(cols_cat):
    plt.subplot(5, 3, i + 1) # Creamos una cuadrícula de 5x3
    ax = sns.countplot(data=df, x=col, hue='Churn')
    
    # Añadimos los totales sobre las barras
    for container in ax.containers:
        ax.bar_label(container, fontsize=10)
        
    plt.title(f'Análisis de {col} vs Churn')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
cols_num = ['Charges.Monthly', 'Charges.Total']

plt.figure(figsize=(15, 5))

for i, col in enumerate(cols_num):
    plt.subplot(1, 2, i + 1)
    sns.kdeplot(data=df, x=col, hue='Churn', fill=True, common_norm=False)
    plt.title(f'Distribución de {col} por Churn')

plt.tight_layout()
plt.show()
```

A continuación se hace un comentaro sobre cada variable en la base de datos:
* Género: No hay diferencia por género
* SeniorCitizen: Porcentualmente, la gente mayor de edad cancela menos.
* Partner: la gente que no es partner cancela más.
* Dependents: La gente sin dependientes cancela más.
* PhoneService: No hay diferencia significativa.
* MultipleLines: No presentan diferencia significativa.
* InternetService: La gente con Fibra óptica cancela mucho más.
* Online Security: La gente sin seguridad cancela más.
* OnlineBackup: La gente sin respaldo cancela más.
* DeviceProtection: La gente sin protección a sus equipos cancela más.
* TechSupport: Las personas sin soporte técnico cancelan más.
* StreamingTV: No hay diferencia.
* StreamingMovies:  No hay diferencia.
* PaperlessBilling: Porcentualmente, cancelan más las personas con pepperlessbilling.
* PymentMethod: Solo el método de pago por cheques electrónicos presenta una tasa de cancelación mucho mayor.
* CHargesMonthly: Los mayores cobros mensuales es donde más cancelaciones hay.
