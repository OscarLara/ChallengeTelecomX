## Extracción
Empezamos por la extracción de datos directamente desde el enlace

```python
# Comenzamos importando todas las librerías necesarias.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Nos conectamos a la fuente de datos directa:
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"
datos = pd.read_json(url)
# Vemos las primeras filas del DataFrame.
datos.head()
```

Nos pudimmos conectar sin problema a los datos originales y cargar un DataFrame.<br>
### Conoce los datos
Podemos ver inmediatamente que las columnas "customer", "phone", "internet" y "account" están anidadas en formato .json. Usaremos .info() para saber más de los datos.

```python
datos.info()
```

Tenemos 7267 registros y 6 columnas de las cuales ninguna tiene valores vacíos y todas son Dtype object. Podríamos empezar a realizar cambios en éste punto, pero lo dejaremos para la siguiente sección. Lo que sí podemos hacer por ahora es verificar que la columna "customerID" no tenga valores repetidos, de esta forma nos aseguramos que cada entrada corresponda a un cliente diferente.

```python
# Verificación de unicidad de customerID. Si el resultado es menor que 7267, hay IDs duplicados.
datos["customerID"].nunique()
```

Ahora podemos crear sub DataFrames para conocer los valores que están anidados.

```python
# Creamos un df por cada columna que es un JSON anidado.
customer = pd.json_normalize(datos["customer"])
phone = pd.json_normalize(datos["phone"])
internet = pd.json_normalize(datos["internet"])
account = pd.json_normalize(datos["account"])
```

## Transformación
Ya que todas nuestras bases están cargadas, podemos empezar a transformar los datos según nos convenga, para ello, primero voy a concatenar la información.

```python
# De los datos originales, solo queremos conservar la columna "Churn"
# Queremos concatenar las tablas horizontalmente, éso es el eje 1.
df = pd.concat([datos["Churn"], customer, phone, internet, account], axis=1)
df.head()
```

```python
# Verificamos la información del nuevo DataFrame.
df.info()
```

Aquí corroboramos que ninguna columna tiene datos nulos. Ahora verficaremos los datos numéricos, si todos están enlistados como tal y si son del tipo correcto.<br>
* SeniorCitizen: int64, es correcto.
* tenure: int64, es correcto.
* Charges.Monthly: float64, es correcto.
* Charges.Total: object, es falso. Hay que cambiar el tipo de dato a float64, ya que es un valor numérico con decimales.

Toca verificar que los datos de cada columna son lo esperado.

```python
# Verificamos si en la columna "Churn", todos los valores son "Yes" o "No".
df["Churn"].unique()
```

Notamos que hay 3 tipos de datos, "No", "Yes" y  ''. Debemos sustituir '' por np.NaN y quitarlo de la base ya que no nos sirve pues no sabesmos si son cancelaciones o no.

```python
# Convertimos nuestros valores vacíos a NaN para poder trabajar con ellos.
df["Churn"] = df["Churn"].replace('', np.nan)
df["Churn"].unique()
```

```python
# Retiramos las filas con valores NaN en la columna "Churn".
df.dropna(subset=["Churn"], inplace=True)
df["Churn"].unique()
```

```python
# Verificamos cuántas entradas nos quedan.
df.info()
```

Nos quedamos con 7043 entradas que sabemos si cancelaron o no su servicio

```python
# Intentamos transformar el dato a float64.
df["Charges.Total"].astype("float64")
```

El error nos indica que hay entradas con valores que no se pueden convertir en numero, en este caso fue un espacio vacío.

```python
df["Charges.Total"].replace(' ', np.nan).astype("float64")
```

```python
# La celda pasada funcionó, por lo que podemos guardar la variable y volvemos a verificar el DataFrame.
df["Charges.Total"] = df["Charges.Total"].replace(' ', np.nan).astype("float64")
df.info()
```

```python
df.head()
```

Nuestro DataFrame está listo para empezar el análisis.

```python
df.describe()
```

A partir de este punto, podemos manipular los datos para conseguir la información que queremos.<br>
Como lo que nos interesa es saber por qué la gente está cancelando, crearemos un DataFrame con solo los datos de las personas que han cancelado.

```python
# Filtramos los valores de las personas que sí cancelaron el servicio.
churn = df[df["Churn"]=="Yes"]
churn.info()
```

Observamos que tenemos 1869 personas que han cancelado de los 7043 datos totales. Ésto representa un $26.53\%$ de la base.<br>
Como ya filtramos las personas que sí cancelaron, podemos eliminar la columna de "Churn".


```python
churn =churn.drop(columns=["Churn"])
churn.head()
```

Notamos que el índice mantiene la numeración previa, como éste es un nuevo DF, le pondremos sus propios índices.

```python
churn = churn.reset_index()
churn = churn.drop(columns=["index"])
churn.head()
```
